# Visualización de datos a través de capas: Altair

## Librerías utilizadas

```python
import pandas as pd
import numpy as np
import altair as alt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

Iniciamos cargando el conjunto de datos de Google Play Store Apps

```python
df = pd.read_csv('https://raw.githubusercontent.com/lihkir/Uninorte/main/AppliedStatisticMS/DataVisualizationRPython/Lectures/Python/PythonDataSets/googleplaystore.csv')
df.head()
```

Eliminamos las entradas del Dataframe con datos NAN.

```python
df_clean = df.dropna()
df_clean.head()
```

### Gráfico de barras del número de aplicaciones en cada categoría Content Rating

```python
content_rating_counts = df_clean['Content Rating'].value_counts().reset_index()
content_rating_counts.columns = ['Content Rating', 'Count']

chart = alt.Chart(content_rating_counts).mark_bar().encode(
    x='Content Rating',
    y='Count',
    color='Content Rating'
).properties(
    width=400, 
    height=400 
)

chart.display()
```

Se observa que el contenido más visto es la categoría 'Everyone' seguido de Teen y Mature 17+.

### Mapa de calor del número de aplicaciones en la app en rangos Category y Rating

```python
df_clean['Rating'] = df_clean['Rating'].round() 
grouped_data = df_clean.groupby(['Category', 'Rating']).size().reset_index(name='Count')

heatmap = alt.Chart(grouped_data).mark_rect().encode(
    x='Category:O', 
    y='Rating:O',   
    color='Count:Q' 
).properties(
    width=600,
    height=300
)

heatmap.display()
```

Se observa que la mayoría de las categorías tiene una distribución de calificaciones que tiende a ser más alta, es decir, más hacia el 4 y 5. Sin embargo, hay ciertas categorías, como "Health and Medical", donde se observa una barra azul en la calificación 4, indicando un número especialmente alto de calificaciones en ese nivel.

Además, algunas categorías tienen una cantidad menor de calificaciones en general, mientras que otras tienen una cantidad más alta de calificaciones, sugiriendo una mayor popularidad o uso de esos productos.

### Combinar ambos gráficos

```python
chart | heatmap
```

Se observan los gráficos anteriores en una misma vista, lo que facilita la comprensión de las variables.
